```python
import pandas as pd
df = pd.read_csv("Top Indian Places to Visit.csv(5.2.24).csv")
df.head()
```

```python
df.isna().sum()
```

```python
df.isna().sum()
```

```python
col = ['Zone','Significance','time needed to visit in hrs','Google review rating','Entrance Fee in INR']
newdf = df[col]
newdf.head()
```

```python
newdf = pd.get_dummies(newdf,dtype = int)
newdf.head()
```

```python
x = newdf.drop(columns = ['Entrance Fee in INR'])
y = newdf['Entrance Fee in INR']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.3)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
from sklearn.linear_model import Lasso
model1 = Lasso(alpha = 20,max_iter = 100,tol = 0.1)
model1.fit(xtrain,ytrain)
```

```python
model1.score(xtrain,ytrain)
```

```python
model1.score(xtest,ytest)
```

```python
from sklearn.linear_model import Ridge
model2 = Ridge(alpha = 25,max_iter=100,tol=0.1)
model2.fit(xtrain,ytrain)
```

```python
model2.score(xtrain,ytrain)
```
